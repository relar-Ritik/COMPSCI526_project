```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./WHO_Suicide_MissingHDIFixed.csv').drop(columns=['Unnamed: 0'])
```

```python
df['gdp_per_capita ($)
```

```python
for key in ['gdp_per_capita ($)','HDI for year']:
    
    y = df['suicides/100k pop']
    X = df[key]
    if key == 'HDI for year':
        val_filter = X!=0
        X = X[val_filter]
        y = y[val_filter]
    plt.scatter(X,y,s = 5,alpha = .4)
    plt.title(f'{key} vs. suicideRate')
    plt.show()
    
for key in ['country','year','sex','age','MissingHDI']:
    ys = []
    
    uniqueVals = sorted(df[key].unique())
    if key == 'age':
        uniqueVals = [uniqueVals[3]]+uniqueVals[0:3]+uniqueVals[-2:]
    
    for value in uniqueVals:
        filtered_ID = df[key] == value
        ys.append(df[filtered_ID]['suicides/100k pop'])
    plt.figure(figsize=(14,7))
    plt.boxplot(ys, tick_labels= uniqueVals, showfliers=False)
    plt.xticks(rotation=-90)
    plt.title(f'{key} vs. suicideRate',)
    plt.show()
    
        
```
